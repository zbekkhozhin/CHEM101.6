```python
dssp=md.compute_dssp(trj,simplified=True)

def dssp_convert(dssp):
 dsspH=np.copy(dssp)
 dsspE=np.copy(dssp)
 dsspH[dsspH=='H']=1                                                                                      
 dsspH[dsspH=='E']=0                                                                                      
 dsspH[dsspH=='C']=0                                                                                      
 dsspH[dsspH=='NA']=0 
 dsspH=dsspH.astype(int)
 TotalH=np.sum(dsspH,axis=1) 
 Hprop=np.sum(dsspH,axis=0).astype(float)/len(dsspE) 

 dsspE[dsspE=='H']=0
 dsspE[dsspE=='E']=1
 dsspE[dsspE=='C']=0
 dsspE[dsspE=='NA']=0
 dsspE=dsspE.astype(int)
 TotalE=np.sum(dsspE,axis=1) 
 Eprop=np.sum(dsspE,axis=0).astype(float)/len(dsspE) 
 return Hprop, Eprop

Hprop,Eprop=dssp_convert(dssp)
plt.plot(residue_num,Hprop,c='r',label='helix')
plt.plot(residue_num,Eprop,c='b',label='sheet')

plt.legend(loc="upper left")
plt.xlabel('Residue', size=18)
plt.ylabel('Secondary Strcture Fraction', size=18)
plt.tick_params(labelsize=18)
```

```python
dssp_full=md.compute_dssp(trj,simplified=False)

def dssp_convert_type(dssp,type):
 dsspS=np.copy(dssp)
 dsspS[dsspS!=type]=0                                                                                      
 dsspS[dsspS==type]=1                                                                                      
 dsspS=dsspS.astype(int)
 TotalS=np.sum(dsspS,axis=1) 
 Sprop=np.sum(dsspS,axis=0).astype(float)/len(dsspS) 
 return Sprop

for t in ['H','E','I','G','B','T','S',' ']:
  Tprop=dssp_convert_type(dssp_full,t)
  plt.plot(residue_num,Tprop,label=t)
  print("SS:",t,np.asarray(residue_num)[np.where(Tprop>.90)])

plt.legend(loc="upper left")
plt.xlabel('Residue', size=18)
plt.ylabel('Secondary Strcture Fraction', size=18)
plt.tick_params(labelsize=18)
```

```python
###Lets Make a Pretty Free Energy Plot of PC1 vs. PC2

def free_energy(a, b, T, y0, ymax, x0, xmax):
    free_energy, xedges, yedges = np.histogram2d(
        a, b, 30, [[y0, ymax], [x0, xmax]], density=True, weights=None)
    free_energy = np.log(np.flipud(free_energy)+.000001)
    free_energy = -(0.001987*T)*free_energy
    return free_energy, xedges, yedges

PC1=reduced_cartesian[:, 0]
PC2=reduced_cartesian[:,1]

dG,xedges,yedges=free_energy(PC2, PC1, 300, -6, 10, -8, 10)

fig = plt.figure(figsize=(8, 6))
im = plt.imshow(dG, interpolation='gaussian', extent=[
                yedges[0], yedges[-1], xedges[0], xedges[-1]], cmap='jet', aspect='auto')
cbar_ticks = [0, 1, 2, 3, 4, 5,6,7,8]
plt.xlim(-8,10)
cb = plt.colorbar(ticks=cbar_ticks, format=('% .1f'),
                  aspect=10)  # grab the Colorbar instance
imaxes = plt.gca()
plt.ylabel("PC2", size=35, labelpad=15)
plt.xlabel("PC1", size=35, labelpad=15)
plt.xticks(fontsize=35)
plt.yticks(fontsize=35)
plt.xticks(size='26')
plt.yticks(size='32')
plt.axes(cb.ax)
plt.clim(vmin=0.01, vmax=8.0)
plt.yticks(size='26')
plt.tight_layout()
```

```python
#2
from scipy.optimize import curve_fit

def exp(x,a,b,c):
    return b*np.exp(-a*x)+c

def autocorr2(x,lags):
    '''numpy.corrcoef, partial'''
    corr=[1. if l==0 else np.corrcoef(x[l:],x[:-l])[0][1] for l in lags]
    c, c_covar = curve_fit(exp,lags,corr)
    if(c[0] != 0):
        c[0] = 1/c[0]
    return c

time=np.linspace(0,1000,trj.n_frames)


lags=range(50)

phi_c = []
decay_phi = []
psi_c = []
decay_psi = []
chi1_c = []
decay_chi1 = []
chi2_c = []
decay_chi2 = []

for i in range(0,77):
    if i in dihedral_dict['phi']:
        phi_c.append(autocorr2(dihedral_dict['phi'][i],lags))
        decay_phi.append(phi_c[i][0])
    else:
        phi_c.append([0,0,0])
        decay_phi.append(0)

    if i in dihedral_dict['psi']:
        psi_c.append(autocorr2(dihedral_dict['psi'][i],lags))
        decay_psi.append(psi_c[i][0])
    else:
        psi_c.append([0,0,0])
        decay_psi.append(0)
    if i in dihedral_dict['chi1']:
        chi1_c.append(autocorr2(dihedral_dict['chi1'][i],lags))
        decay_chi1.append(chi1_c[i][0])
    else:
        chi1_c.append([0,0,0])
        decay_chi1.append(0)
    if i in dihedral_dict['chi2']:
        chi2_c.append(autocorr2(dihedral_dict['chi2'][i],lags))
        decay_chi2.append(chi2_c[i][0])
    else:
        chi2_c.append([0,0,0])
        decay_chi2.append(0)

threshold = 5

decay_phi_np = np.array(decay_phi)
decay_psi_np = np.array(decay_psi)
decay_chi1_np = np.array(decay_chi1)
decay_chi2_np = np.array(decay_chi2)

high_phis=np.where(decay_phi_np > threshold)
high_psis=np.where(decay_psi_np > threshold) 
high_chi1s=np.where(decay_chi1_np > threshold)
high_chi2s=np.where(decay_chi2_np > threshold)

print("Following residues have high correlation time for their:\n\nPhis:",high_phis,"\n\nPsis:",high_psis,"\n\nChi1s:",high_chi1s,"\n\nChi2s:",high_chi2s,"\n\n")

#C1C2bondlength[:,0]
phi40_acf=autocorr1(dihedral_dict['phi'][40],lags)
phi60_acf=autocorr1(dihedral_dict['phi'][60],lags)
psi8_acf=autocorr1(dihedral_dict['psi'][8],lags)
psi30_acf=autocorr1(dihedral_dict['psi'][30],lags)
psi39_acf=autocorr1(dihedral_dict['psi'][39],lags)
psi51_acf=autocorr1(dihedral_dict['psi'][51],lags)
chi2_59_acf=autocorr1(dihedral_dict['chi2'][59],lags)

plt.plot(time[lags],phi40_acf,color='b',label='phi_40')
#plt.plot(time[lags],phi_c[39][1]*np.exp(-phi_c[39][0]*time[lags])+phi_c[39][2],color='b',label='phi_40_fit')
plt.plot(time[lags],phi60_acf,color='r',label='phi_60')
#plt.plot(time[lags],phi_c[59][1]*np.exp(-phi_c[59][0]*time[lags])+phi_c[59][2],color='r',label='phi_60_fit')
plt.plot(time[lags],psi8_acf,color='g',label='psi_8')
#plt.plot(time[lags],psi_c[7][1]*np.exp(-psi_c[7][0]*time[lags])+psi_c[7][2],color='g',label='psi_8_fit')
plt.plot(time[lags],psi30_acf,color='c',label='psi_30')
#plt.plot(time[lags],psi_c[29][1]*np.exp(-psi_c[29][0]*time[lags])+psi_c[29][2],color='c',label='psi_30_fit')
plt.plot(time[lags],psi39_acf,color='m',label='psi_39')
#plt.plot(time[lags],psi_c[3][1]*np.exp(-psi_c[38][0]*time[lags])+psi_c[38][2],color='m',label='psi_39_fit')
plt.plot(time[lags],psi51_acf,color='y',label='psi_51')
#plt.plot(time[lags],psi_c[50][1]*np.exp(-psi_c[50][0]*time[lags])+psi_c[50][2],color='y',label='psi_51_fit')
plt.plot(time[lags],chi2_59_acf,color='k',label='chi2_59')
#plt.plot(time[lags],chi2_c[58][1]*np.exp(-chi2_c[58][0]*time[lags])+chi2_c[58][2],color='k',label='chi2_59_fit')

plt.legend()
plt.xlabel("Time ($\mu$s)", size=14, labelpad=15)
plt.ylabel("Autocorrelation", size=14, labelpad=15)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xticks(size='14')
plt.yticks(size='14')
plt.show()
```
